# 상관관계 분석 후 히트맵 그리기

```python
# 라이브러리 임포트

# Data
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

#warning
import warnings
warnings.filterwarnings('ignore')
```

```python
# 엑셀 파일 불러오기

df = pd.read_csv('E:/GSR/light_data/CSVDAT~2/var_김수기.csv')
df = df.drop(['frame'] ,axis=1)


# 상관분석 
# dataframe의 corr()함수는 column별로 상관분석을 수행하기 때문에 transpose를 취한 후 상관분석 진행
df_t = df.T
corr_t = df_t.corr(method = 'spearman')
# corr_t = df_t.corr(method = 'pearson')
corr_t

# 첫 번째 column만 가져오기
corr_1 = corr_t[0]
df_corr1 = pd.DataFrame(corr_1)
df_corr1

# (225, 1)을 (15, 15)로 변환
df_array = df_corr1.to_numpy()  # 데이터프레임을 배열로 넘파이로 변환

arr_1 = df_array[1:16].T
arr_2 = df_array[16:31].T
arr_3 = df_array[31:46].T
arr_4 = df_array[46:61].T
arr_5 = df_array[61:76].T
arr_6 = df_array[76:91].T
arr_7 = df_array[91:106].T
arr_8 = df_array[106:121].T
arr_9 = df_array[121:136].T
arr_10 = df_array[136:151].T
arr_11 = df_array[151:166].T
arr_12 = df_array[166:181].T
arr_13 = df_array[181:196].T
arr_14 = df_array[196:211].T
arr_15 = df_array[211:226].T

arr_res = np.concatenate([arr_1, arr_2, arr_3, arr_4, arr_5, 
                          arr_6, arr_7, arr_8, arr_9, arr_10,
                          arr_11, arr_12, arr_13, arr_14, arr_15], axis=0)
# print(np.shape(arr_res))
```

```python
# 시각화
df_res = pd.DataFrame(arr_res)
fig = plt.figure(figsize=(10, 10))
sns.heatmap(df_res,
            annot = True,      # 실제 값 화면에 나타내기
            cmap = 'RdYlBu_r',  # Red, Yellow, Blue 색상으로 표시
            vmin = -1, vmax = 1, #컬러차트 -1 ~ 1 범위로 표시
           )
plt.show()

fig.savefig('김수기_var_s.png', orientation="landscape")  # 이미지 저장
```

```python
# 시각화
df_res = pd.DataFrame(arr_res)
fig = plt.figure(figsize=(10, 10))
sns.heatmap(df_res,
            annot = True,      # 실제 값 화면에 나타내기
            cmap = 'YlGnBu',  # Yellow, Green, Blue 색상으로 표시
            vmin = -1, vmax = 1, #컬러차트 -1 ~ 1 범위로 표시
           )
plt.show()
```
