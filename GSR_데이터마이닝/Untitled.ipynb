# Roi_capture

```python
import cv2

file_name = ['불러올 파일 이름.mp4']

# 지정할 영역의 범위
start_points = [(270, 210)]
end_points = [(310, 260)]

for i, name in enumerate(file_name):
    cap = cv2.VideoCapture(name)
    frame_len = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    print(i, ' : ', frame_len)
    for j in range(int(frame_len)):
        ret, frame = cap.read()
        if not ret:
            break

        cv2.rectangle(frame, start_points[i], end_points[i], 1, 1, 1)

        cv2.imshow('frame', frame)

        key = cv2.waitKey(30)

        if key == ord('q') or key == ord(' '):
            break

        elif key == ord('s'):
            cv2.imwrite('지정 영역 확인용 캡쳐 코드.png', frame)

    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()
```

# Roi_csv

```python
import cv2
import numpy as np
import csv

# 지정한 영역의 범위
start_points = [(270, 210)]
end_points = [(310, 260)]

file_name = ['불러올 파일 이름.mp4']

for i, name in enumerate(file_name):
    cap = cv2.VideoCapture(name)
    frame_len = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    print(i, ': ', frame_len)

    # 저장 파일 이름
    f = open(name[:-4] + '.csv', 'w', newline='')
    writer = csv.writer(f)
    for j in range(int(frame_len)):
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        crop = gray[start_points[i][1]:end_points[i][1], start_points[i][0]:end_points[i][0]]

        # 평균 밝기값
        mean = np.mean(crop)
        # 표준편차
        std = np.std(crop)

        writer.writerow([j, mean, std])

    #         cv2.rectangle(frame, start_points[i], end_points[i], 1, 1, 1)

    #         cv2.imshow('Crop', crop)
    #         cv2.imshow('Frame', frame)
    #         key = cv2.waitKey(1)
    #         if key == ord('q') or key == ord(' '):
    #             break
    #         elif key == ord('s'):
    #             cv2.imwrite('a.png', frame)
    f.close()
#     if key == ord('q'):
#         break
cv2.destroyAllWindows()
```
